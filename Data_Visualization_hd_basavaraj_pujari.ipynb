```python
#Task 1: Bar Plots and Pie Charts
#a. Visualize Categorical Variables
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv(r'C:\Users\HC-WIN\Downloads\heart_disease_health_indicators_BRFSS2015 2 (1).csv')

# Set the style for the plots
sns.set(style='whitegrid')

# Pie Chart for Smoking Status
plt.figure(figsize=(8, 6))
data['Smoker'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Distribution of Smokers')
plt.ylabel('')
plt.show()
```

```python
# Bar Plot for Physical Activity
plt.figure(figsize=(8, 6))
sns.countplot(x='PhysActivity', data=data, palette='pastel')
plt.title('Physical Activity Status')
plt.xlabel('Physically Active (0=No, 1=Yes)')
plt.ylabel('Count')
plt.show()
```

```python
#b. Compare Heart Disease Across Categories
```

```python
data['HeartDiseaseorAttack'] = data['HeartDiseaseorAttack'].astype(str)

# Grouped Bar Chart for Heart Disease by Smoking Status
plt.figure(figsize=(10, 6))
sns.countplot(x='Smoker', hue='HeartDiseaseorAttack', data=data, palette='pastel')
plt.title('Heart Disease Prevalence by Smoking Status')
plt.xlabel('Smoker (0=No, 1=Yes)')
plt.ylabel('Count')
plt.legend(title='Heart Disease', loc='upper right', labels=['0', '1'])  # Specify labels as needed
plt.show()
```

```python
#Task 2: Explore Other Visualization Techniques
```

```python
# Pie Chart for Heart Disease Proportion by Smoking Status
smoker_heart_disease = data.groupby(['Smoker', 'HeartDiseaseorAttack']).size().unstack()
smoker_heart_disease.plot.pie(y=1, autopct='%1.1f%%', startangle=90, legend=False, figsize=(8, 8))
plt.title('Proportion of Heart Disease by Smoking Status')
plt.ylabel('')
plt.show()
```

```python
# Box Plot for BMI by Smoking Status
plt.figure(figsize=(10, 6))
sns.boxplot(x='Smoker', y='BMI', data=data, palette='pastel')
plt.title('BMI Distribution by Smoking Status')
plt.xlabel('Smoker (0=No, 1=Yes)')
plt.ylabel('BMI')
plt.show()
```

```python
# Heatmap for Correlation Matrix
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Health Metrics')
plt.show()
```

```python
# Facet Grid for Heart Disease by Age and Smoking Status
g = sns.FacetGrid(data, col='Smoker', hue='HeartDiseaseorAttack', height=5, aspect=1)
g.map(sns.histplot, 'Age', bins=10, kde=True)
g.add_legend()
g.set_titles(col_template='Smoker: {col_name} (0=No, 1=Yes)')
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Age Distribution by Smoking Status and Heart Disease')
plt.show()
```

```python
# Pair Plot for Selected Features
sns.pairplot(data, vars=['BMI', 'Age', 'HeartDiseaseorAttack'], hue='HeartDiseaseorAttack', palette='pastel')
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```

```python
# Violin Plot for Physical Activity by Heart Disease
plt.figure(figsize=(10, 6))
sns.violinplot(x='HeartDiseaseorAttack', y='PhysActivity', data=data, palette='pastel')
plt.title('Physical Activity Distribution by Heart Disease')
plt.xlabel('Heart Disease (0=No, 1=Yes)')
plt.ylabel('Physical Activity (0=No, 1=Yes)')
plt.show()
```
