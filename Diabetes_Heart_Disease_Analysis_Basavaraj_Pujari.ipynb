```python
# Age Distribution
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
# Age distribution
# Assuming 'Age' is a column in your dataframe 'df'
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=10, kde=False, color='skyblue')

# Add titles and labels
plt.title('Age Distribution of Participants', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Number of Participants', fontsize=12)

# Show plot
plt.show()
```

```python
# BMI distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['BMI'], bins=20, kde=False, color='lightgreen')

# Add titles and labels
plt.title('BMI Distribution of Participants', fontsize=16)
plt.xlabel('BMI', fontsize=12)
plt.ylabel('Number of Participants', fontsize=12)

# Show plot
plt.show()
```

```python
# Mental health days distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='MentHlth', data=df, palette='coolwarm')

# Add titles and labels
plt.title('Frequency of Mental Health Days', fontsize=16)
plt.xlabel('Number of Mental Health Days', fontsize=12)
plt.ylabel('Count of Participants', fontsize=12)

# Show plot
plt.show()
```

```python
# Physical health days distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='PhysHlth', data=df, palette='Blues')

# Add titles and labels
plt.title('Frequency of Physical Health Days', fontsize=16)
plt.xlabel('Number of Physical Health Days', fontsize=12)
plt.ylabel('Count of Participants', fontsize=12)

# Show plot
plt.show()
```

```python
#Investigate Prevalence of Health Conditions
# Prevalence of High Blood Pressure
high_bp_counts = df['HighBP'].value_counts()

# Labels for the pie chart
labels = ['No High Blood Pressure', 'High Blood Pressure']

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(high_bp_counts, labels=labels, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90, explode=(0, 0.1))

# Add a title
plt.title('Prevalence of High Blood Pressure', fontsize=16)

# Show the chart
plt.show()
```

```python
# Prevalence of High Cholesterol
high_chol_counts = df['HighChol'].value_counts()

# Labels for the pie chart
labels = ['No High Cholesterol', 'High Cholesterol']

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(high_chol_counts, labels=labels, autopct='%1.1f%%', colors=['lightgreen', 'salmon'], startangle=90, explode=(0, 0.1))

# Add a title
plt.title('Prevalence of High Cholesterol', fontsize=16)

# Show the chart
plt.show()
```

```python
# Cholesterol check prevalence
chol_check_counts = df['CholCheck'].value_counts()

# Labels for the pie chart
labels = ['No Cholesterol Check', 'Cholesterol Check']

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(chol_check_counts, labels=labels, autopct='%1.1f%%', colors=['lightcoral', 'skyblue'], startangle=90, explode=(0, 0.1))

# Add a title
plt.title('Percentage of Individuals Who Had a Cholesterol Check', fontsize=16)

# Show the chart
plt.show()
```

```python
# Prevalence of Diabetes
diabetes_counts = df['Diabetes'].value_counts()

# Set up the bar chart
plt.figure(figsize=(8, 6))
plt.bar(diabetes_counts.index, diabetes_counts.values, color=['lightgreen', 'orange'])

# Add titles and labels
plt.title('Prevalence of Diabetes', fontsize=16)
plt.xlabel('Diabetes Status', fontsize=14)
plt.ylabel('Number of Participants', fontsize=14)
plt.xticks(ticks=diabetes_counts.index, labels=['No Diabetes', 'Diabetes'])

# Show the chart
plt.show()
```

```python
# Prevalence of Smoking
smoker_counts = df['Smoker'].value_counts()

# Labels for the pie chart
labels = ['Non-Smoker', 'Smoker']

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(smoker_counts, labels=labels, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90, explode=(0, 0.1))

# Add a title
plt.title('Smoking Status of Participants', fontsize=16)

# Show the chart
plt.show()
```

```python
#Analyze Distribution of Heart Disease (Target Variable)
# Heart Disease distribution bar chart
heart_disease_counts = df['HeartDiseaseorAttack'].value_counts()

# Set up the bar chart
plt.figure(figsize=(8, 6))
plt.bar(heart_disease_counts.index, heart_disease_counts.values, color=['lightcoral', 'skyblue'])

# Add titles and labels
plt.title('Distribution of Heart Disease', fontsize=16)
plt.xlabel('Heart Disease Status', fontsize=14)
plt.ylabel('Number of Participants', fontsize=14)
plt.xticks(ticks=heart_disease_counts.index, labels=['No Heart Disease', 'Heart Disease'])

# Show the chart
plt.show()
```

```python
df['HeartDiseaseorAttack'] = df['HeartDiseaseorAttack'].astype(str)
df['HighBP'] = df['HighBP'].astype(str)
df['Smoker'] = df['Smoker'].astype(str)
```

```python
#Explore Relationships with Heart Disease
#Age vs. Heart Disease
import warnings 
warnings.filterwarnings('ignore')
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Age', hue='HeartDiseaseorAttack', palette='pastel')
plt.title('Age vs. Heart Disease Prevalence', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of Cases', fontsize=14)
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert HeartDiseaseorAttack to integer if it's not already
df['HeartDiseaseorAttack'] = df['HeartDiseaseorAttack'].astype(int)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            linewidths=.5, cbar=True, square=True, vmin=-1, vmax=1)

# Add titles and labels
plt.title('Correlation Matrix', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.tight_layout()

# Show the heatmap
plt.show()
```

```python
#  analyze the correlation between BMI and HeartDiseaseorAttack
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='BMI', y='HeartDiseaseorAttack', alpha=0.7)
plt.title('BMI vs. Heart Disease', fontsize=16)
plt.xlabel('BMI', fontsize=14)
plt.ylabel('Heart Disease (0 = No, 1 = Yes)', fontsize=14)
plt.axhline(0.5, color='red', linestyle='--')  # Add a line at the y=0.5 for reference
plt.tight_layout()
plt.show()
```

```python
#C. Compare Heart Disease Across Demographic Groups
#High Blood Pressure vs. Heart Disease
# Calculate the counts of heart disease based on high blood pressure
high_bp_hd_counts = df.groupby(['HighBP', 'HeartDiseaseorAttack']).size().unstack()

# Plotting the bar chart
plt.figure(figsize=(8, 6))
high_bp_hd_counts.plot(kind='bar', stacked=True, color=['lightblue', 'salmon'])
plt.title('High Blood Pressure vs. Heart Disease', fontsize=16)
plt.xlabel('High Blood Pressure (0 = No, 1 = Yes)', fontsize=14)
plt.ylabel('Number of Individuals', fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
#Smoking Status vs. Heart Disease
# Calculate the counts of heart disease based on smoking status
smoker_hd_counts = df.groupby(['Smoker', 'HeartDiseaseorAttack']).size().unstack()

# Plotting the bar chart
plt.figure(figsize=(8, 6))
smoker_hd_counts.plot(kind='bar', stacked=True, color=['lightgreen', 'orange'])
plt.title('Smoking Status vs. Heart Disease', fontsize=16)
plt.xlabel('Smoking Status (0 = Non-Smoker, 1 = Smoker)', fontsize=14)
plt.ylabel('Number of Individuals', fontsize=14)
plt.xticks(rotation=0)
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```
